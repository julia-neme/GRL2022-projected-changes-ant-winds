```python
import dask.distributed as dsk
import glob
import xarray as xr
import xesmf as xe

import warnings # ignore these warnings
warnings.filterwarnings("ignore", category = FutureWarning)
warnings.filterwarnings("ignore", category = UserWarning)
warnings.filterwarnings("ignore", category = RuntimeWarning)

clnt = dsk.Client()
clnt
```

# CMIP6

```python
era = xr.open_dataset('psl-monthly-1979_2015-erai.nc')
path_uas = glob.glob('cmip6/uas*')

for i in range(len(path_uas)):
    
    model_name = path_uas[i].split('/')[-1].split('_')[-1].split('.')[0]
    
    ds = xr.open_dataset(path_uas[i])
    ds_in = ds['uas'].to_dataset()
    ds_out = era['uas'].to_dataset()
    
    regrid = xe.Regridder(ds_in, ds_out, 'conservative')
    dr_out = regrid(ds_in)
    dr_out.to_netcdf('cmip6-regridded/uas_historical_rg_'+model_name+'.nc')
```
